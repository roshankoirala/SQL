# Operators

```sql
SELECT col_name FROM tab_name 
WHERE NOT [CONDITION_1] 
    AND [CONDITION_2]
    OR [CONDITION_3];
```


__Comparison operator__

```sql
SELECT col_1 FROM tab_name 
WHERE (col_1 = [VAL_1])
    AND (col_2 > [VAL_2])
    OR (col_3 <= [VAL_3])
    AND (col_4 <> [VAL_4]); 
```

# Conditional 

__Simple case statement__

```sql
SELECT col_1
, CASE @col_2 
    WHEN [CONDITION_1] THEN [VAL_1] 
    WHEN [CONDITION_2] THEN [VAL_2]
    ELSE [VAL_3]
  END AS col_2_cat 
FROM tab_name; 
```
	 
__Nested case statement__

```sql
SELECT col_1
, CASE @col_2 
    WHEN [CONDITION_1] THEN [VAL_1] 
    ELSE CASE
        WHEN [CONDITION_21] THEN [VAL_21]
        WHEN [CONDITION_22] THEN CASE
            WHEN [CONDITION_221] THEN [VAL_221]
            ELSE [VAL_222]
        END 
        ELSE [VAL_23]
    END 
  END AS col_2_cat 
FROM tab_name; 
```

# Table Joins

__Two ways to join tables__

```sql
SELECT * 
FROM tab_1 AS t1 
JOIN tab_2 AS t2 
    ON t1.key_1 = t2.key_2;
```


```sql
SELECT * 
FROM tab_1 AS t1 
JOIN tab_2 AS t2 
    WHERE t1.key_1 = t2.key_2;
```

__INNER JOIN__ $ A \cap B $

```sql
SELECT * 
FROM tab_1 AS t1 
INNER JOIN tab_2 AS t2 
    ON t1.key_1 = t2.key_2;
```

__LEFT JOIN__ $A \cup (A \cap B )$ 

```sql
SELECT * 
FROM tab_1 AS t1 
LEFT JOIN tab_2 AS t2 
    ON t1.key_1 = t2.key_2;
```

__RIGHT JOIN__ $ (A \cap B) \cup B $ 

```sql
SELECT * 
FROM tab_1 AS t1 
RIGHT JOIN tab_2 AS t2 
    ON t1.key_1 = t2.key_2;
```


__FULL (OUTER) JOIN__ $ A \cup B $

```sql
SELECT * 
FROM tab_1 AS t1 
FULL JOIN tab_2 AS t2 
    ON t1.key_1 = t2.key_2;
```



__CROSS JOIN__ $ A \times B $

```sql
SELECT t1.col_1, t2.col_2 
FROM tab_1 AS t1 
CROSS JOIN tab_2 AS t2 -- no join condition 
```

```sql
SELECT *, (CASE 
    WHEN col_1 > tab_avg.col_1_avg THEN 'class A'
    ELSE 'Class B' 
END) AS classes 
FROM tab_name 
CROSS JOIN (
    SELECT AVG(col_1) AS col_1_avg FROM tab_name
) AS tab_avg 
```


__SELF JOIN__ $ A \circ A $

```sql
SELECT parent.name, child.name 
FROM tab_name AS parent 
JOIN tab_name AS child 
    ON parent.id = child.parent_id; 
```


```sql
SELECT parent.name, child.name 
FROM tab_name AS parent 
LEFT JOIN tab_name AS child 
    ON parent.id = child.parent_id; 
```

# Sub-Queries
__Scalar__ 

```sql
SELECT * FROM tab_name 
WHERE col_1 > (
    SELECT AVG(col_1) FROM tab_name 
); 
```


```sql
SELECT * FROM tab_name AS t 
JOIN (SELECT AVG(col_1) AS col_1_avg FROM tab_name) AS t_avg
    ON t.col_1 > t_avg.col_1_avg; 
```

__Multi row__


```sql
SELECT * FROM tab_name 
WHERE (col_1, col_2) IN (
    SELECT col_1, AVG(col_2) FROM tab_name 
    GROUP BY col_1 
);
```


```sql
SELECT * FROM tab_1 AS t1 
WHERE t1.col_1 IN (
    SELECT DISTINCT t2.col_1 FROM tab_2 AS t2
    WHERE [CONDITION]
); 
```



__Correlated__ 

```sql
SELECT * FROM tab_name AS t
WHERE col_1 > (
    SELECT AVG(col_1) FROM tab_name 
    WHERE col_2 = t.col_2 
);
```

```sql
SELECT * FROM tab_1 AS t
WHERE col_1 > (
    SELECT AVG(col_1) FROM tab_2 
    WHERE col_2 = t.col_2 
);
```

Tips: avoid correlated sub-query if possible.

__ALL / ANY / EXISTS__

```sql
SELECT * FROM tab_name 
WHERE [cond_1]
    AND col_1 > ALL (
        SELECT col_2 FROM tab_name 
    )
    OR col_2 < ANY (
        SELECT col_3 FROM tab_name 
    );
```


```sql
SELECT * FROM tab_name 
WHERE EXIST (
        SELECT 1 FROM tab_name 
        WHERE [CONDITION_1]
    )
    AND NOT EXIST (
        SELECT 1 FROM tab_name 
        WHERE [CONDITION_2]
    );
```

Tips: `ALL` can potentially be avoided using `MAX`. 

Tips: `EXISTS` can potentially be avoided using `IN`.


__Nested Subquery__

```sql
SELECT * FROM (
    SELECT col_1, SUM(col_2) AS col_2_total 
    FROM tab_name 
    GROUP BY col_1
) AS tab_total 
JOIN (
    SELECT AVG(col_2_total) AS col_2_avg 
        FROM (
            SELECT col_1, SUM(col_2) AS col_2_total 
            FROM tab_name 
            GROUP BY col_1 
        ) AS tab_avg 
)
ON tab_total.col_2_total > tab_avg.col_2_avg; 
```

Tips: Nested subquery can can be cleaner using `WITH` clause 


__Subquery with HAVING__

```sql
SELECT col_1, AVG(col_2) FROM tab_name 
GROUP BY col_1 
HAVING AVG(col_2) > (
    SELECT AVG(col_2) FROM tab_name 
); 
```

# Window functions
	

__With usual aggrigate__

```sql
SELECT col_1, col_2 
, AVG(col_1) OVER (
    PARTITION BY col_2  
) AS avg1_part2  
FROM tab_name 
```

__RANK_NUMBER, RANK, DENSE_RANK__

`ROW_NUMBER{A, B, B, C} -> {1, 2, 3, 4}`

`RANK{A, B, B, C} -> {1, 2, 2, 4}`

`DENSE_RANK{A, B, B, C} -> {1, 2, 2, 3}`


```sql
SELECT col_1, col_2, col_3  
, ROW_NUMBER() OVER (
    PARTITION BY col_2 
    ORDER BY col_3 
) AS row_num
, RANK() OVER (
    PARTITION BY col_2 
    ORDER BY col_3 
) AS rank
, DENSE_RANK() OVER (
    PARTITION BY col_2 
    ORDER BY col_3 
) AS dense_rank
FROM tab_name
```

Alternatively 

```sql
SELECT col_1, col_2, col_3  
, ROW_NUMBER() OVER w AS row_num
, RANK() OVER w AS rank
, DENSE_RANK() OVER w AS dense_rank
FROM tab_name
WHERE [CONDITION]
WINDOW w AS (
    PARTITION BY col_2 
    ORDER BY col_3 
)
```


Select top-10 from each group 

```sql
SELECT * FROM (
    SELECT col_1, col_2, col3
    , RANK() OVER (
        PARTITION BY col_1 
        ORDER BY col_2 DESC 
    ) AS rank 
    FROM tab_name 
) AS x 
WHERE x.rank <= 10;
```



__LEAD, LAG__

`LAG{A, B, C, D} -> {NULL, A, B, C}`

`LEAD{A, B, C, D} -> {B, C, D, NULL}`


```sql
SELECT *
, LAG(col_1, [STEP_LAG], [DEFAULT_VAL]) OVER w AS lag_col_1 
, LEAD(col_1, [STEP_LEAD], [DEFAULT_VAL]) OVER w AS lead_col_1 
FROM tab_name 
WINDOW w AS (
    PARTITION BY col_2 
    ORDER BY col_3 
    [RANGE BETWEEN UNBOUNDED PRECEDING
        AND CURRENT FOLLOWING] -- default frame 
); 
```




__FIRST_VALUE, LAST_VALUE, NTH_VALUE__

```sql
SELECT *
, FIRST_VALUE(col_1) OVER w AS col_1_first 
, LAST_VALUE(col_1) OVER w AS col_1_last 
FROM tab_name
WINDOW w AS (
    PARTITION BY col_2 
    ORDER BY col_3 
    RANGE BETWEEN UNBOUNDED PRECEDING
        AND UNBOUNDED FOLLOWING -- frame for last value 
); 
```


```sql
SELECT *
, NTH_VALUE(col_1, [N]) OVER w AS col_1_nth 
, NTH_VALUE(col_1, -[N]) OVER w AS col_1_nth_last 
FROM tab_name
WINDOW w AS (
    PARTITION BY col_2 
    ORDER BY col_3 DESC 
    ROWS BETWEEN 5 PRECEDING
        AND 5 FOLLOWING -- custom frame 
); 
```

__NTILE, CUME_DIST, PERCENT_RANK__

```sql
SELECT t.col_1, t.col_2 
, CASE 
    WHEN t.n_buckets_by_col2 = 1 THEN 'cat_1'
    WHEN t.n_buckets_by_col2 = 2 THEN 'cat_2'
    ...
    ELSE 'cat_n'
END AS cat_col2 
FROM (
    SELECT *
    , NTILE([N]) OVER (
        PARTITION BY col_1
        ORDER BY col_2 
    ) AS n_buckets_by_col2
    FROM tab_name 
    WHERE [CONDITION]
) AS t; 
```


Pick top N% of the data 

```sql
SELECT t.col_1, t.col_2 
FROM (
    SELECT *
    , CUM_DIST() OVER (
        PARTITION BY col_1
        ORDER BY col_2 DESC 
    ) cum_dist_col2_part1 
    FROM tab_name 
) AS t 
WHERE t.cum_dist_col2_part1 <= [N]/100
```


What's the rank of col_1 = 'VAL_1'?

```sql
SELECT t.rank_col1_by_col2 
FROM (
    SELECT *
    , PERCENT_RANK() OVER (
        PARTITION BY col_1
        ORDER BY col_2 DESC 
    ) AS rank_col1_by_col2 
    FROM tab_name 
) AS t 
WHERE t.col_1 = 'VAL_1'; 
```
